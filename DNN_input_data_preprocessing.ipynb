```python
import csv
import pandas as pd
import numpy as np
```

```python
all_trip_data = pd.read_csv('tripad_data.csv')

all_trip_data#.head()
```

```python
UPC_data = pd.read_csv('UPC.csv')

del UPC_data['Unnamed: 0']

author = UPC_data['Author']

author
```

```python
UPC_data = all_trip_data[all_trip_data['Author'].isin(author)]
UPC_data = UPC_data.sort_values(by=['Author'],axis=0)
UPC_data = UPC_data.reset_index()
del UPC_data['index']
del UPC_data['Unnamed: 0']

UPC_data

UPC_data.to_csv("DNN_input_data.csv")
```
